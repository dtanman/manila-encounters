## Functions

```python
def clean(story):

    # remove URLs
    story = re.sub(
        '(http|ftp|https):\/\/[\w\-_]+(\.[\w\-_]+)+([\w\-\.,@?^=%&amp;:/~\+#]*[\w\-\@?^=%&amp;/~\+#])?',
        '',
        story
    )
    
    # remove them ellipses
    story = re.sub('â€¦', '', story)

    # remove picture uploads
    story = re.sub('pic.twitter.com/\\w+\\b', '', story)

    # remove hashtags
    story = re.sub('#\\w+\\b', '', story)

    # remove mentions
    story = re.sub('@\\w+\\b', '', story)

    # retain only alphanumeric, dashes
    # https://stackoverflow.com/questions/3210393/how-do-i-remove-all-non-alphanumeric-characters-from-a-string-except-dash
    story = re.sub('[^a-zA-Z0-9 -]', '', story)

    # all lowercase letters
    story = story.lower()

    # spacing
    story = ' '.join(story.split())
    
    # remove stopwords
    story = ' '.join([x for x in story.split() if x not in stops])
    
    return story
```

### With other snippets

```python
def find_n_lsi(snippet, n=5, return_the='nothing'):
    story = clean(snippet)
    temp_bow = dictionary.doc2bow(snippet.split())
    q = index[lsi[temp_bow]]
    best_n = sorted(enumerate(q), key=lambda item: -item[1])[:n]

    if return_the=='best_n':
        return best_n
    
    if return_the=='df':
        return df.iloc[[x[0] for x in best_n]][['tweet_id', 'text', 'clean_text', 'cleaner_text', 'cleanest_text', 'cleanester_text', 'date', 'time', 'likes', 'retweets']]
    
    with pd.option_context('display.max_colwidth', -1):
        display(df.iloc[[x[0] for x in best_n]][['tweet_id', 'text', 'clean_text', 'cleaner_text', 'cleanest_text', 'cleanester_text', 'date', 'time', 'likes', 'retweets']])
```

## Bigrams

We use a `Phrases` object from `gensim` to help us with creating bigrams, instead of just mixing and mashing all pairs of words.

The `Phrases` object is a model -- it will first infer if the the bigram is statistically significant in the corpus, before including it into the corpus.

### Prep

```python
phrases = Phrases([x.split() for x in df['cleanester_text']], min_count=10)
bigram = Phraser(phrases)

df['final_text'] = [' '.join(bigram[x.split()]) for x in df['cleanester_text']]
```

```python
with pd.option_context('display.max_colwidth', -1):
    display(df.sort_values('retweets', ascending=False)[['text', 'cleanester_text', 'final_text']])
```

```python
with pd.option_context('display.max_colwidth', -1):
    display(
        df[
            (df['clean_text'].str.contains('getting')) &
            (df['clean_text'].str.contains('hotter'))
        ].sort_values('retweets', ascending=False)[[
            'tweet_id', 'text', 'clean_text', 'cleanester_text', 'final_text'
        ]]
    )
```

```python
with pd.option_context('display.max_colwidth', -1):
    display(
        df[
            (df['clean_text'].str.contains('end')) &
            (df['clean_text'].str.contains('2012'))
        ].sort_values('retweets', ascending=False)[[
            'tweet_id', 'text', 'clean_text', 'cleanester_text', 'final_text'
        ]]
    )
```

To create trigrams, we can just reuse the same `bigram` method!

```python
phrases = Phrases([x.split() for x in df['final_text']], min_count=10)
bigram = Phraser(phrases)

df['final_text'] = [' '.join(bigram[x.split()]) for x in df['final_text']]
```

### Finalization

Let's redo the corpus and LSI models to contain the new bigrams.

```python
# redo tokenization and dictionary (needs to be recreated
# because of new bigrams and trigrams)
tokenized = [nltk.word_tokenize(x) for x in df['final_text']]
dictionary = corpora.Dictionary(tokenized)
dictionary.save('../gensim/tweets_3.dict')

# redo corpus
corpus = [dictionary.doc2bow(text) for text in tokenized]
tfidf = models.TfidfModel(corpus)
corpus_tfidf = tfidf[corpus]

# redo LSI model
lsi = models.LsiModel(corpus_tfidf, id2word=dictionary, num_topics=20)
corpus_lsi = lsi[corpus_tfidf]
```
